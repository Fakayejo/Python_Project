```python
# Select rows where 'Exited' equals 1 (churned customers)
customer_churn[customer_churn["Exited"]==1]
```

```python
# Update the 'CreditScore' column by adding 10 points to each value
customer_churn["CreditScore"]+10
```

```python
# Add a new column 'CreditScore_Categories' based on the 'CreditScore' values (Using a function)

def Credit_Score_Category(Scores):
    if Scores < 300:
        return "Bad"
    elif Scores < 600:
        return "poor"
    elif Scores < 700:
        return "Fair"
    elif Scores < 800:
        return "Good"
    else:
        return "Excellent"
    
customer_churn["CreditScore_Categories"] = customer_churn["CreditScore"].apply(Credit_Score_Category)
customer_churn
```

```python
customer_churn.head()
```

```python
# Drop the 'RowNumber' column
customer_churn.drop(1,axis = 0)
```

```python
# Saving the modified DataFrame as a new CSV
customer_churn.to_csv("Modified_cus_churn.csv", index = False)
```
